```python
import requests
import bs4
import html2text
import time
```

# to scrap over links and append them in the array -->> the_links

```python
start = time.time()
the_links=['https://gate.ahram.org.eg/','https://www.youm7.com/','https://www.alarabiya.net/'] #the start page
the_count=0 #page number to scrape
count=1
while len(the_links)<2481:
    #print('####################### New page #################################',the_count)
    res=requests.get(the_links[the_count])#res.text
    the_count+=1
    soup=bs4.BeautifulSoup(res.text,'lxml')
    for link in soup.find_all('a',href=True):
        if  str(link['href']).startswith('https://') and str(link['href']) not in the_links and count<2481:
            print('<<<<<<<<<<<<<<<<link number>>>>>>>>>>',count)
            the_links.append(link['href'])
            print(link['href'])
            count+=1
```

# next code to add all links in the text file

```python
links_count=0 
with open('Downloads/files of search engine project/thefile555.txt', 'a', encoding="utf-8") as file:
    for link in the_links:
            file.write("{}\n".format(link))
            #links_count+=1
```

# next code to make the relations between the links
## note: it take more than 3 min for 1000 links so it should run in colab

```python
    with open('Downloads/files of search engine project/thefile555.txt', 'a', encoding="utf-8") as file:
        for i_of_l in range(len(the_links)):
            print('###### New page ######',i_of_l)
            try:
                res3=requests.get(the_links[i_of_l])#res.text
                soup3=bs4.BeautifulSoup(res3.text,'lxml')
                for l in soup3.find_all('a',href=True):
                    if  str(l['href']).startswith('https://') and str(l['href']) in the_links:
                        file.write("{} {}\n".format(i_of_l,the_links.index(l['href'])))
                        links_count+=1
            except: print("Invalid Link")
```

# that fun made to add the number of links and the number of relations in the beginning of the file

```python
def line_prepender(filename, line):
    with open(filename, 'r+') as f:
        content = f.read()
        f.seek(0, 0)
        f.write(line.rstrip('\r\n') + '\n' + content)
        
line_prepender('Downloads/files of search engine project/thefile.txt','{} {}'.format(len(the_links),links_count-len(the_links)))
end = time.time()
```
