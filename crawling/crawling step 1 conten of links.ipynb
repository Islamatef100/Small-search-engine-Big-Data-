```python
import requests
import bs4
import html2text
```

```python
the_links=['https://gate.ahram.org.eg/'] #the start page
the_count=0 #page number to scrape
count=1
while len(the_links)<10000:
    print('####################### New page #################################',the_count)
    res=requests.get(the_links[the_count])#res.text
    the_count+=1
    soup=bs4.BeautifulSoup(res.text,'lxml')
    for link in soup.find_all('a',href=True):
        if  str(link['href']).startswith('https://') and str(link['href']) not in the_links and count<10000:
            print('<<<<<<<<<<<<<<<<link number>>>>>>>>>>',count)
            the_links.append(link['href'])
            print(link['href'])
            count+=1
        
```
